```python
import numpy as np


def under_attack(col, queen):
    return col in queen or \
           any(abs(col - x) == len(queen) - i for i, x in enumerate(queen))


def solve(n):
    solutions = [[]]
    for row in range(n):
        solutions = [solution + [i + 1]
                     for solution in solutions
                     for i in range(n)
                     if not under_attack(i + 1, solution)]
    return solutions

def queens(size):
    result = np.array(solve(size))
    for answer in result:
        print("------------------------")
        print("------------------------")
        print("ABCDEFGH")
        print("".join(map(str,answer)))
        print("ABCDEFGH")
        print("------------------------")
        print("------------------------")
queens(8)


def highest_comb(L):
    max_length = len(str(max(L)))
    L = map(str, L)
    def sort_key(s):
        return s + s[-1] * (max_length - len(s))

    sorted_list = sorted(L, key=sort_key, reverse=True)
    return ''.join(sorted_list)


print(highest_comb([50, 9, 2, 1]))             # 95021
print(highest_comb([90, 9]))                   # 990
print(highest_comb([115, 15, 7, 905, 904]))    # 905904715115
```
