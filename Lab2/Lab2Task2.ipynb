```python

import  random



strArray = ["apple", "cat", "dog", "one", "two", "three", "sun", "moon", "rise", "mobo", "cat", "cat", "cat", "cat"]

print("Обычное рехеширование  : ")

def hash(astring, tablesize):
    sum = 0
    for pos in range(len(astring)):
        sum = sum + ord(astring[pos])

    return sum % tablesize

def ReHashFunc(pos, i, size):

    return (pos + i) % size

hashTable = [None] *len(strArray)
size = len(strArray)

for j in range(len(strArray)):

    i = 0

    if hashTable[hash(strArray[j],size)] == None:

        hashTable[hash(strArray[j], size)] = strArray[j]

    else:

        while hashTable[ReHashFunc(hash(strArray[j], size), i, size)] != None:

            i += 1

        hashTable[ReHashFunc(hash(strArray[j], size), i, size)] = strArray[j]

print(hashTable)




print("Хеширование псевдорандомом : ")



hashTableRandom = [None]*len(strArray)


def hashRandom(astring, tablesize):
    sum = 0
    for pos in range(len(astring)):
        sum = sum + ord(astring[pos]) * 7

    return sum % tablesize


for j in range(len(strArray)):
    i = 0
    if hashTableRandom[hashRandom(strArray[j],size)] == None:
        hashTableRandom[hashRandom(strArray[j], size)] = strArray[j]

    else:

        while hashTableRandom[ReHashFunc(hashRandom(strArray[j], size), i, size)] != None:

            i = random.randint(1,size)

        hashTableRandom[ReHashFunc(hash(strArray[j], size), i, size)] = strArray[j]
print(hashTableRandom)





print("Хеширование цепочкой : ")

hashTableNode = [None] * len(strArray)

for i in range(len(strArray)):

    if hashTableNode[hash(strArray[i], len(strArray))] == None:
        hashTableNode[hash(strArray[i], len(strArray))] = strArray[i]
    elif hashTableNode[hash(strArray[i], len(strArray))] != None:
        prevValue = hashTableNode[hash(strArray[i], len(strArray))]
        if type(prevValue) != "list":
            hashTableNode[hash(strArray[i], len(strArray))] = []
            if type(prevValue) is list:
                for j in range(len(prevValue)):
                    hashTableNode[hash(strArray[i], len(strArray))].append(prevValue[j])
            else:
                hashTableNode[hash(strArray[i], len(strArray))].append(prevValue)
        hashTableNode[hash(strArray[i], len(strArray))].append(strArray[i])


for i in range(len(hashTableNode)) :
    if i == None:
        hashTableNode.remove(None)

print(hashTableNode)


def findValue(key, array):
    for idxI, itemI in enumerate(array):
        if (type(itemI) is list):
            for idxJ, itemJ in enumerate(array[idxI]):
                if (itemJ == key):
                    return f'{idxI} - {itemJ} ({idxJ})'
        else:
            if (itemI == key):
                return f'{idxI} - {itemI}'
    return array[hash(key, len(strArray))]


print(findValue("three", hashTableNode))
print(findValue("cat",hashTableRandom))
print(findValue("dog",hashTable))







```
